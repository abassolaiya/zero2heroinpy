```python
for x in sorted(pytz.country_names):
    print("{} : {}". format(x, pytz.country_names[x]), end=': ')
    if x in pytz.country_timezones:
          print(pytz.country_timezones[x])
    else:
          print('Time zone not defined')
```

```python
#to print time zones with the time in each zone
for x in sorted(pytz.country_names):
    print("{} : {}". format(x, pytz.country_names[x]), end=': ')
    if x in pytz.country_timezones:
        for zone in sorted(pytz.country_timezones[x]):
            tz_here = pytz.timezone(zone)
            local_time = datetime.datetime.now(tz=tz_here)
            print('\t\t {} : {}'.format(zone, local_time))
    else:
          print('\t\t Time zone not defined')
```

```python
# Always work in UTC and convert to local time when you're displaying that time to the used
#Daylight saving wahala
local_time = datetime.datetime.now()
utc_time = datetime.datetime.utcnow()

print ("Native local time {}".format(local_time))
print("Native_UTC{}".format(utc_time))

aware_local_time = pytz.utc.localize(utc_time).astimezone()
aware_utc_time = pytz.utc.localize(utc_time)

print("Aware local time(), time zone {}".format(aware_local_time, aware_utc_time.tzinfo))
print("Aware_UTC {}, time_zone {}".format(aware_utc_time, aware_utc_time.tzinfo))
```
