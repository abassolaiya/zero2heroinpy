```python
#Packing widget inside a frame can help make the postioning a little bit better
mainWindow = tkinter.Tk()

mainWindow.title('Learning Python GUI')
mainWindow.geometry('640x480+8+400')

label = tkinter.Label(mainWindow, text="Learning python GUI")
label.pack(side='top')

leftFrame = tkinter.Frame(mainWindow)
leftFrame.pack(side='left', anchor='n', fill=tkinter.Y, expand=False)

canvas = tkinter.Canvas(leftFrame, relief='raised', borderwidth=1)
canvas.pack(side='left', anchor='n')

rightFrame = tkinter.Frame(mainWindow)
rightFrame.pack(side='right', anchor='n', expand=False)

buttonA = tkinter.Button(rightFrame, text='ButtonA')
buttonB = tkinter.Button(rightFrame, text='ButtonB')
buttonC = tkinter.Button(rightFrame, text='ButtonC')

buttonA.pack(side='top')
buttonB.pack(side='top')
buttonC.pack(side='top')
mainWindow.mainloop()
```

```python
#Pack is useful for simple layouts and is the simplest of the three geometry managers
#Now using Grid Geometry manager

mainWindow = tkinter.Tk()

mainWindow.title('Learning Python GUI')
mainWindow.geometry('640x480-8-200')

label = tkinter.Label(mainWindow, text="Learning python GUI")
label.grid(row=0, column=0)

leftFrame = tkinter.Frame(mainWindow)
leftFrame.grid(row=1, column=1)

canvas = tkinter.Canvas(leftFrame, relief='raised', borderwidth=1)
canvas.grid(row=1, column=0)

rightFrame = tkinter.Frame(mainWindow)
rightFrame.grid(row=1, column=2)

buttonA = tkinter.Button(rightFrame, text='ButtonA')
buttonB = tkinter.Button(rightFrame, text='ButtonB')
buttonC = tkinter.Button(rightFrame, text='ButtonC')

buttonA.grid(row=0, column=0)
buttonB.grid(row=1, column=0)
buttonC.grid(row=2, column=0)
mainWindow.mainloop()
```

```python
#Adding sticky property for the anchor points of the buttons' position

mainWindow = tkinter.Tk()

mainWindow.title('Learning Python GUI')
mainWindow.geometry('640x480-8-200')

label = tkinter.Label(mainWindow, text="Learning python GUI")
label.grid(row=0, column=0)

leftFrame = tkinter.Frame(mainWindow)
leftFrame.grid(row=1, column=1)

canvas = tkinter.Canvas(leftFrame, relief='raised', borderwidth=1)
canvas.grid(row=1, column=0)

rightFrame = tkinter.Frame(mainWindow)
rightFrame.grid(row=1, column=2, sticky='n')

buttonA = tkinter.Button(rightFrame, text='ButtonA')
buttonB = tkinter.Button(rightFrame, text='ButtonB')
buttonC = tkinter.Button(rightFrame, text='ButtonC')

buttonA.grid(row=0, column=0)
buttonB.grid(row=1, column=0)
buttonC.grid(row=2, column=0)
mainWindow.mainloop()
```

```python
#Adding sticky property for the anchor points of the buttons' position

mainWindow = tkinter.Tk()

mainWindow.title('Learning Python GUI')
mainWindow.geometry('640x480-8-200')

label = tkinter.Label(mainWindow, text="Learning python GUI")
label.grid(row=0, column=0)

leftFrame = tkinter.Frame(mainWindow)
leftFrame.grid(row=1, column=1)

canvas = tkinter.Canvas(leftFrame, relief='raised', borderwidth=1)
canvas.grid(row=1, column=0)

rightFrame = tkinter.Frame(mainWindow)
rightFrame.grid(row=1, column=2, sticky='n')

buttonA = tkinter.Button(rightFrame, text='ButtonA')
buttonB = tkinter.Button(rightFrame, text='ButtonB')
buttonC = tkinter.Button(rightFrame, text='ButtonC')

buttonA.grid(row=0, column=0)
buttonB.grid(row=1, column=0)
buttonC.grid(row=2, column=0)

#column configure to give columns weight
mainWindow.columnconfigure(0, weight=1)
mainWindow.columnconfigure(1, weight=1)
mainWindow.grid_columnconfigure(2, weight=1 )
mainWindow.mainloop()
```

```python
#Adding sticky property for the anchor points of the buttons' position

mainWindow = tkinter.Tk()

mainWindow.title('Learning Python GUI')
mainWindow.geometry('640x480-8-200')

label = tkinter.Label(mainWindow, text="Learning python GUI")
label.grid(row=0, column=0)

leftFrame = tkinter.Frame(mainWindow)
leftFrame.grid(row=1, column=1)

canvas = tkinter.Canvas(leftFrame, relief='raised', borderwidth=1)
canvas.grid(row=1, column=0)

rightFrame = tkinter.Frame(mainWindow)
rightFrame.grid(row=1, column=2, sticky='n')

buttonA = tkinter.Button(rightFrame, text='ButtonA')
buttonB = tkinter.Button(rightFrame, text='ButtonB')
buttonC = tkinter.Button(rightFrame, text='ButtonC')

buttonA.grid(row=0, column=0)
buttonB.grid(row=1, column=0)
buttonC.grid(row=2, column=0)

#column configure to give columns weight
mainWindow.columnconfigure(0, weight=1)
mainWindow.columnconfigure(1, weight=1)
mainWindow.grid_columnconfigure(2, weight=1 )

leftFrame.config(relief='sunken', borderwidth=1)
rightFrame.config(relief='sunken', borderwidth=1)
leftFrame.grid(sticky='ns')
rightFrame.grid(sticky='new')
mainWindow.mainloop()
```
